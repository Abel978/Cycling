```python
import numpy as np
from env import State, OneStepEnv
from Agent import Single_State_Agent
```

```python
def ablation_train(p0, e0):
    S = State(p0, e0)
    env = OneStepEnv(S)

    for alpha in [0.01, 0.1, 0.3, 0.5]:
        A = Single_State_Agent(S, env, alpha=alpha)

        A.train(num_episodes=200000, verbose=False)

        print(f"Agent with alpha={alpha} -- {A.Q}")
    # Do the initial beliefs madder?

    for initial_beliefs in [0, -1, 1, 5]:
        A = Single_State_Agent(S, env, alpha=0.1)

        A.train(num_episodes=200000, verbose=False)

        print(f"Agent with alpha={alpha} -- {A.Q}")
```

```python
# 2b

p0 = np.array([[0], [0,0], [2]], dtype=object)
e0 = np.array([[3], [2,2], [0]], dtype=object)

ablation_train(p0, e0)
```
